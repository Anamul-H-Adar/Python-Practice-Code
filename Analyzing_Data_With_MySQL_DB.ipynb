```python
# Executing Second cursor
cur1.execute('SELECT * FROM sample_db.orderdetails')
```

```python
# Fetch Data
orderdtl_df = cur1.fetchall()
```

```python
# Fetch data into Dataframe
orderdtl_df = pd.DataFrame(orderdtl_df)
```

```python
# Define Columns Name
order_dtl_col = ["orderNumber","productCode","quantity","price","orderLineNumber"] 

# Set Columns Name
orderdtl_df.columns = order_dtl_col
```

```python
# Merge two dataframe or inner join two dataframe
total_order = pd.merge(order_df, orderdtl_df, on = 'orderNumber', how = 'inner')
```
