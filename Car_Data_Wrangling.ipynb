### Identifying and handle missing values

```python
df.head(5)
```

```python
import numpy as np
# Replace '?' to NaN 
df.replace('?', np.nan, inplace = True)
```

```python
df.head()
```

```python
missing_data = df.isnull()
```

```python
missing_data.head(5)
```

```python
# Count missing values for each column
# missing_data['normalized-losses'].value_counts()

for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

### Deal with missing data

```python
# Numerical observation should replace with mean value
mean_normalized_losses = df['normalized-losses'].astype('float').mean()
df['normalized-losses'].replace(np.nan, mean_normalized_losses, inplace = True)
```

```python
mean_bore = df['bore'].astype('float').mean()
df['bore'].replace(np.nan, mean_bore, inplace = True)
```

```python
mean_stroke = df['stroke'].astype('float').mean()
df['stroke'].replace(np.nan, mean_stroke, inplace = True)
```

```python
mean_horsepower = df['horsepower'].astype('float').mean()
df['horsepower'].replace(np.nan, mean_horsepower, inplace = True)
```

```python
mean_peak_rpm = df['peak-rmp'].astype('float').mean()
df['peak-rmp'].replace(np.nan, mean_peak_rpm, inplace = True)
```

```python
# Replace categorical value with most frequent value
df['num-of-doors'].value_counts()
```

```python
df['num-of-doors'].value_counts().idxmax()
```

```python
df['num-of-doors'].replace(np.nan, 'four', inplace = True)
```

```python
# Delete entire row where price column value is null
df.dropna(subset = ['price'], inplace = True)
```

```python
# Becase we delete two rows
df.reset_index(drop = True, inplace = True)
```

### Correct Data Format

```python
# Convert data types to proper format
df['bore'] = df['bore'].astype('float') # format can change single

# format can change multiple at a time when datatype will same
df[['stroke', 'price', 'peak-rmp']] = df[['stroke', 'price', 'peak-rmp']].astype('float')
df['normalized-losses'] = df['normalized-losses'].astype('int')
```

```python
df.dtypes
```

### Standardzation
#### Standardization is the process of transforming data into a common format

```python
# 1 mile per gallon is equal to 235.214583 liters per 100 km.
df['city_l/100'] = 235/df['city-mpg']
```

```python
df['highway_l/100'] = 235/df['highway-mpg']
```

```python
df.head()
```

### Data Normalization
#### Normalization is the process of transforming values of several variables into a similar range.
#### would like to normalize those variables so their value ranges from 0 to 1

```python
# scale the columns "length", "width" and "height".
# replace original value by (original value)/(maximum value)
df['length'] = df['length']/ df['length'].max()
df['width'] = df['width']/ df['width'].max()
df['height'] = df['height'] / df['height'].max()
```

```python
df.head()
```

### Binning
#### Binning is a process of transforming continuous numerical variables into discrete categorical 'bins' for grouped analysis.

```python
# We will use the pandas method 'cut' to segment the 'horsepower' column into 3 bins.
df['horsepower'] = df['horsepower'].astype('int')
```

```python
# Let's plot the histogram of horsepower to see what the distribution of horsepower looks like.
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot

plt.pyplot.hist(df['horsepower'])
plt.pyplot.xlabel('Hoursepower')
plt.pyplot.ylabel('Count')
plt.pyplot.title('Hoursepower bins')
```

```python
bins = np.linspace(df['horsepower'].min(), df['horsepower'].max(), 4)
```

```python
bins
```

```python
# Group name
group_name = ['Low','Medium','High']
```

```python
df['hoursepower_binned'] = pd.cut(df['horsepower'], bins, labels = group_name, include_lowest = True)
```

```python
df[['horsepower', 'hoursepower_binned']].head()
```

```python
df['hoursepower_binned'].value_counts()
```

```python
plt.pyplot.bar(group_name,df['hoursepower_binned'].value_counts())
plt.pyplot.xlabel('Hoursepower')
plt.pyplot.ylabel('Count')
plt.pyplot.title('Hoursepower Bins')
```

```python
plt.pyplot.hist(df['horsepower'], bins = 3)
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```
